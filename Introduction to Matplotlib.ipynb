## Two options for subplots

```python
# subplot option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize = (10,5))

ax1.plot(x, x/2);
ax2.scatter(np.random.randn(10) , np.random.randn(10));

# and so on....
```

```python
## Subplot option 2

fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize = (10,5))

## Plot to different indexes

ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.randn(10) , np.random.randn(10));

# and so on...
```

## Plotting from pandas dataframe

```python
import pandas as pd
```

```python
# make a dataframe

car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
## Replace extra character sin price column with regex

car_sales["Price"] = car_sales["Price"].str.replace('[\$\ , \.]' , '')
car_sales
```

```python
## Remove last two zeroes

car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020" , periods = len(car_sales))
car_sales
```

```python
## Cumilative sum of all the prices

car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
## Plot total sales

car_sales.plot(x="Sale Date" , y="Total Sales");
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
```

```python
## Scatter plot

car_sales.plot(x = "Odometer (KM)" , y = "Price" , kind = "scatter");
```

```python
## Bar plot

car_sales.plot(x = "Odometer (KM)" , y = "Price" , kind = "bar");
```

```python
# Histogram

car_sales["Odometer (KM)"].plot.hist(bins=25);
```

## New Dataset

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize = (10,30) , subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Scatter plot of age Vs cholrestrol

## PyPlot

over_50.plot(kind = "scatter", 
            x= "age",
            y ="chol",
            c = "target");
```

```python
## OO Method mixed with pyplot

fig, ax = plt.subplots(figsize = (10,6))

over_50.plot(kind = "scatter", 
            x= "age",
            y ="chol",
            c = "target",
            ax = ax);

# Adjust the X-axis

# ax.set_xlim([45,90]);
```

## Object Oriented Method from scratch

```python
# OO Method from scratch

# Set the style
plt.style.use("seaborn-whitegrid");

fig, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"],
                    cmap = "winter")

#Customize the plot
ax.set(title = "Heart disease and cholestrol levels",
      xlabel = "Age",
      ylabel = "Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title = "target");

# Add a horizontal line
ax.axhline(y = over_50["chol"].mean() , linestyle = "--");
```

```python
over_50.head()
```

## Plot a subplot for age, chol and thalach

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize = (10,10), sharex = True)

# Add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"],
                     cmap = "winter")

#Customize the plot
ax0.set(title = "Heart disease and cholestrol levels",
      ylabel = "Cholestrol");

# Change x axis limit
ax0.set_xlim([50,80])

# Add a legend
ax0.legend(*scatter.legend_elements(), title = "target");

# Add a horizontal line
ax0.axhline(y = over_50["chol"].mean() , linestyle = "--");


# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                    y = over_50["thalach"],
                    c = over_50["target"], 
                      cmap="winter")

#Customize the plot
ax1.set(title = "Heart disease and Max heart rate",
      xlabel = "Age",
      ylabel = "Thalach");

# Change x axis limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend
ax1.legend(*scatter.legend_elements(), title = "target");

# Add a horizontal line
ax1.axhline(y = over_50["thalach"].mean() , linestyle = "--");

# Add title to the plot

fig.suptitle("Heart Disease Analysis" , fontsize = 16, fontweight = "bold");
```
